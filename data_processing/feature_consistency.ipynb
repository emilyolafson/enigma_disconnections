```python
import numpy as np
from plot_workbench import * 
import plot_workbench
from imp import reload
import seaborn as sns
import matplotlib.pyplot as plt
reload(plot_workbench)


dist = '/home/ubuntu/enigma/results/analysis_1/shen268_normed_motor_scores_chacovol_acutechronic_ridge_crossval1_median_betas_allperms_95.txt'
dist = '/home/ubuntu/enigma/results/analysis_1/betas_allperms_shen.npy'
temp =np.load(dist)
final_w = np.loadtxt('/home/ubuntu/enigma/results/analysis_1/final_model_weights_alldata.txt')

temp_mean = np.mean(temp, axis=1)
temp_std = np.std(temp,axis=1)
reshaped = np.reshape(temp, [500,268])

logic = np.count_nonzero(reshaped,axis=0)>475
std_onlyfreq = temp_std*logic
mean_onlyfreq = temp_mean*logic
median_onlyfreq_mean = np.median(mean_onlyfreq,axis=0)
std_onlyfreq_mean = np.mean(std_onlyfreq,axis=0)

print(np.max(median_onlyfreq_mean))
print(np.min(median_onlyfreq_mean))
print(np.count_nonzero(std_onlyfreq_mean))
print(np.max(std_onlyfreq_mean))
print(np.min(std_onlyfreq_mean))

np.savetxt('/home/ubuntu/enigma/results/mean_onlyfreq.txt',median_onlyfreq_mean)
np.savetxt('/home/ubuntu/enigma/results/std_onlyfreq.txt',std_onlyfreq_mean)
```
