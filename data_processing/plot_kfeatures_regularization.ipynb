```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

results_path = '/home/ubuntu/enigma/results'
output_path = 'analysis_1'
atlas = 'shen268'
y_var = 'normed_motor_scores'
chaco_type = 'chacovol'
subset = 'acutechronic'
model_tested = 'ridge'
crossval = '1'
rootname = os.path.join(results_path, output_path, '{}_{}_{}_{}_{}_crossval{}'.format(atlas, y_var, chaco_type, subset, model_tested,crossval))

alphas = np.empty(shape=(100,5))
nfeats = np.empty(shape=(100,5))

for n in range(0,100):
    alphas[n,] = np.loadtxt(rootname +'_perm'+ str(n)  +'_alphas.txt')
    nfeats[n,] = np.loadtxt(rootname +'_perm'+ str(n)  +'_nfeats.txt')

alpha = np.unique(alphas, return_counts=True)
nfeat = np.unique(nfeats, return_counts=True)

alphas = np.reshape(alphas, [500,])
nfeats = np.reshape(nfeats, [500,])


df = pd.DataFrame({'Lambda': alphas, 'Number of features selected': nfeats})
counts = df.groupby(['Lambda', 'Number of features selected']).size().reset_index(name='Count')
sns.set(font_scale=1.5)
plt.figure(figsize=(8,8))
sns.set_theme(style="whitegrid", palette="pastel")

sns.scatterplot(data=counts, x="Lambda", y="Number of features selected", hue="Count",size="Count")
print(os.path.join(rootname + 'nfeats_lambdas.png'))
plt.savefig(os.path.join(rootname + 'nfeats_lambdas.png'),bbox_inches='tight', dpi=300)
```
